## **Ejercicio 1** (40%)

Construya un DataFrame a partir del siguiente diccionario

`exam_data = {'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas'], 'score': [12.5, 9, 16.5, 10, 9, 20, 14.5, 12, 8, 19], 'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1], 'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']}`

con las siguientes etiquetas para los índices

`labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']`

1. Seleccione solo las columnas *name* y *score*.
  
2. Filtre las filas donde el número de intentos (attemps) sea mayor que 2.

3. Filtre las filas donde el *score* se encuentre entre 15 y 20 (incluidos)

4. Cambie el *score* en la fila 'd' a 11.5

5. Agregue una nueva fila con etiqueta 'k' al dataframe con un valor que usted desee para cada columna

6. Borre la fila 'k' y entregue nuevamente el DataFrame original.

7. Ordene el DataFrame por nombre en orden ascendente.

8. Reemplace los valores de la columna *qualify* que contiene los valores *yes* y *no* por los valores booleanos True y False respectivamente.

```python
import pandas as pd

# Diccionario de datos
exam_data = {
    'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas'],
    'score': [12.5, 9, 16.5, 10, 9, 20, 14.5, 12, 8, 19],
    'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
    'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']
}

# Etiquetas para los índices
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']

# Construir DataFrame
df = pd.DataFrame(exam_data, index=labels)

# Mostrar DataFrame
print(df)
```

```python
# seleccionamos las columnas name y score para esto utilizamos la notación de corchetes para acceder a las dos columnas específicas del df
selected_columns= df[['name','score']]
print(selected_columns)
```

```python
#Filtramos las filas donde el número de intentos (attemps) sea mayor que 2
#Utilizamos loc para acceder a un grupo de filas a través de etiquetas o matriz booleana

filtered_attemps= df.loc[df['attempts'] > 2]
print(filtered_attemps)

#df.loc[df['attempts'] > 2] utiliza esta serie booleana como una máscara para seleccionar las filas donde la condición sea verdadera
```

```python
#Filtramos las filas donde el score se encuentre entre 15 y 20 (incluidos)
#volvemos a utilizar una máscara booleana para seleccionar las filas donde la condición sea verdadera (esta vez la condición conlleva un & y se utiliza el operador = para que sea incluyente)
filtered_score= df.loc[(df['score']>=15)&(df['score']<=20)]
print(filtered_score)
```

```python
#Cambiamos el score en la fila 'd' a 11.5
#Utilizamos loc para acceder a una fila específica por su índice y luego asignamos el valor dado en la columna score
df.loc['d','score']=11.5
print(df)
```

```python
# Agregamos una nueva fila con etiqueta 'k' al dataframe con un valor deseado para cada columna
df.loc['k'] = ['Lucia', 15.0, 2, 'yes']
print(df)
```

```python
#Borramos la fila 'k' con el método drop de pandas y entregamos nuevamente el DataFrame original
df= df.drop('k')
print(df)
```

```python
# Ordenamos el DataFrame por nombre en orden ascendente mediante el método sort_values que por defecto tiene el parámetro ascending en True
df_sorted = df.sort_values(by='name')
print(df_sorted)
```

```python
#Reemplazamos los valores de la columna qualify que contiene los valores yes y no por los valores booleanos True y False respectivamente
#Creamos un diccionario de mapeo de valores
#En pandas el mapeo se utiliza para cambiar los valores en una serie según un diccionario o una función
mapping = {'yes': True, 'no': False}

# Aplicamos el mapeo a la columna 'qualify'
df['qualify'] = df['qualify'].map(mapping)
print(df)
```

## **Ejercicio 2** (30%)

Importe el dataset `athletes` del repositorio del curso y realice las siguientes tareas:

1. Importarlo como nombre athletes
2. Filtre el dataframe tal que solo se quede con los atletas que hayan ganado al menos una medalla.
3. Calcule la masa muscular por fila con una función como

$$I_{m.m.} = \frac{Weight}{Height^2}$$

4. Agrupe los datos por calculando la media por deporte para cada una del resto de características. ¿Cuál es el deporte en el que los jugadores tienen mayor peso promedio?

5. ¿Cuál es el pais con más participación femenina?

```python
import pandas as pd
```

```python
nombre_atlethes = pd.read_csv('/content/athletes.csv')
nombre_atlethes.head(2)
```

```python
# Filtro en el que al menos haya tenido una medalla de oro, plata o bronze
filtro = (nombre_atlethes['gold'] >= 1) | nombre_atlethes['silver'] >= 1 | (nombre_atlethes['bronze'] >= 1)
nombre_atlethes = nombre_atlethes.loc[filtro]
nombre_atlethes.sample(5)
```

```python
def masa_muscular(height, weight):
  """
  Funcion que permite calcular la masa muscula apartir del peso y la altura

  params:
  height(float): Altura
  weight(float): Peso

  Return
  Masa corporal
  """
  masam = weight/(height**2)
  return masam

# Crea una nueva columna `masa_corporal`
nombre_atlethes['masa_corporal'] = nombre_atlethes.apply(lambda row: masa_muscular(row['height'],
                                                                                   row['weight']),
                                                         axis=1)
# Devuelve una muestra
nombre_atlethes.sample(3)
```

```python
# Calculo de peso promedio por deporte
media_peso_deporte = nombre_atlethes.groupby('sport')['weight'].mean().sort_values(ascending= False)
media_peso_deporte.head(3) # Imprime una muestra

# Deporte con jugadores con mayor peso promedio
media_peso_deporte.idxmax()
```

```python
# Filtra las atletas mujeres y agrupa por nacionalidad. Obtiene el país con más mujeres
nombre_atlethes[nombre_atlethes.sex == 'female'].groupby('nationality')['sex'].agg('count').idxmax()
```

## **Ejercicio 3 (30%)**

Importe el dataset de nba del repositorio de clase y responda (hint: Elimine las instancias con registros faltantes):

1. Cuáles son los equipos que pagan los salarios más altos (presente el resultado como un dataFrame con el nombre del equipo y la cantidad de dinero de la nómina)
2. Separe la variable *Age* en tres grupos y cree una nueva columna llamada *Age-Group* con esta información.<
3. Construya una tabla dinámica que muestre el salario promedio de los jugadores por equipo y posición para para los cinco equipos que pagan mejores salarios.

```python
import pandas as pd
```

```python
#Importamos los datos que vamos a necesitar y los definimos como la data
data=pd.read_csv("https://raw.githubusercontent.com/tomasate/Datos_Clases/main/Datos_1/nba.csv")
data.head(10)
```

```python
#Eliminamos las filas que tienen registros faltantes en cualquiera de las columnas
df= data.dropna()
df
```

```python
#Mostramos los 10 salarios mas altos
df= df.sort_values('Salary', ascending=False)
topsalario=df.loc[:,['Team','Salary']].head(10)
topsalario
```

```python
#Ahora veremos como se comportan de acuerdo a los rangos de edad.
df["Age_Group"]=pd.cut(df.Age, bins = 3)
df
```

```python
#para crear la tabla dinamica con promedio de salarios por equipo y posición, extraemos los cinco equipos que mejor salario pagan.
#Ordenamos los salarios

df.sort_values(by='Salary', ascending=False, inplace=True)

# Tomar los cinco primeros equipos
equipos_top5 = df['Team'].head(5).tolist()
equipos_top5
```

```python
#Filtramos el dataframe original por los equipos que están dentro del top 5 y creamos la tabla dinámica que nos permite conocer el salario
#promedio por equipo y posición

tabla_dinamica = df[df['Team'].isin(equipos_top5)].pivot_table(
    index=['Team', 'Position'],
    values='Salary',
    aggfunc='mean')

tabla_dinamica
```
